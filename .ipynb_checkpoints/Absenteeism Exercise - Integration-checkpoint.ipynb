```python
from absenteeism_module import *
model = absenteeism_model('model', 'scaler')
model.load_and_clean_data('Absenteeism_new_data.csv')
df_new_obs = model.predicted_outputs()
df_new_obs
```

# MySQL - Python connection

```python
import pymysql
```

```python
# connect to MySQL database - modify this when productize
# modify user and password for database access

conn = pymysql.connect(database = 'predicted_outputs', user = '', password = '')
```

```python
# create cursor to write MySQL queries in Python IDE
cursor = conn.cursor()
```

```python
# check table in MySQL first
select_query = "select * from predicted_outputs;"
cursor.execute(select_query)
```

```python
# insert data from df_new_obs to predicted_outputs in MySQL

#insert query follows this format insert into ___ values (record), in which record = (value, value),
# and don't forget to end query with ; 
insert_query = 'insert into predicted_outputs values '
records_new_obs = df_new_obs.to_records(index=False)
records_new_obs
```

```python
for record in records_new_obs: 
    row = str(record)
    insert_query += row + ','

#remove the last , after the loop operation
insert_query = insert_query[:-1] + ";"
insert_query
```

```python
cursor.execute(insert_query)
```

```python
conn.commit()
```

```python
# now that we have loaded the data into the table, let's close the connection 
conn.close()
```
